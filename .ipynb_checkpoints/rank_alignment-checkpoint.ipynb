```python
import pandas as pd
from scipy.stats import spearmanr, kendalltau

file_path = 'mcg_paper_table.csv'
df = pd.read_csv(file_path)

categories = ['humanities', 'socialsciences', 'stem', 'others', 'avg']
orig_prefix = 'mmlu_original_'
dgen_prefix = 'mmlu_dgen_'

all_results = []

# Process across all models (ignoring 0-shot or 5-shot distinction)
results = []
for cat in categories:
    col_orig = f'{orig_prefix}{cat}_accuracy'
    col_dgen = f'{dgen_prefix}{cat}_accuracy'

    # (1) Rank calculation (descending)
    df[f'{cat}_Orig_Rank'] = df[col_orig].rank(ascending=False)
    df[f'{cat}_DGEN_Rank'] = df[col_dgen].rank(ascending=False)

    # (2) Spearman correlation
    spear_corr, spear_p = spearmanr(
        df[f'{cat}_Orig_Rank'],
        df[f'{cat}_DGEN_Rank']
    )
    
    # (3) Kendall's Tau
    kend_corr, kend_p = kendalltau(
        df[f'{cat}_Orig_Rank'],
        df[f'{cat}_DGEN_Rank']
    )

    results.append([cat, spear_corr, spear_p, kend_corr, kend_p])

corr_df = pd.DataFrame(
    results,
    columns=["Category", "SpearmanR", "SpearmanP", "KendallTau", "KendallP"]
)
corr_df
```

```python
import matplotlib.pyplot as plt

categories = ['overall'] + [cat for cat in ['humanities', 'socialsciences', 'stem', 'others', 'avg'] if cat != 'avg']

for cat in categories:
    rank_orig_col = f"{cat}_Orig_Rank"
    rank_dgen_col = f"{cat}_DGEN_Rank"
    
    col_orig = f"mmlu_original_{cat}_accuracy" if cat != 'overall' else "mmlu_original_avg_accuracy"
    col_dgen = f"mmlu_dgen_{cat}_accuracy" if cat != 'overall' else "mmlu_dgen_avg_accuracy"
    
    
    if rank_orig_col not in df.columns:
        df[rank_orig_col] = df[col_orig].rank(ascending=False)
    if rank_dgen_col not in df.columns:
        df[rank_dgen_col] = df[col_dgen].rank(ascending=False)

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(11, 11))  
axes = axes.flatten()  

# Plot each category
for i, cat in enumerate(categories):
    rank_orig_col = f"{cat}_Orig_Rank"
    rank_dgen_col = f"{cat}_DGEN_Rank"
    
    if rank_orig_col not in df.columns or rank_dgen_col not in df.columns:
        continue

    ax = axes[i]
    title = "OVERALL" if cat == "overall" else cat.upper()
    ax.scatter(df[rank_orig_col], df[rank_dgen_col], s=60, alpha=0.7)
    
    ax.set_title(f"{title} \n MMLU Original vs. D-GEN Ranks")
    ax.set_xlabel("MMLU Original Rank (1 = best)")
    ax.set_ylabel("MMLU D-GEN Rank (1 = best)")
    
    # Add diagonal line (perfect alignment line)
    min_rank = min(df[rank_orig_col].min(), df[rank_dgen_col].min())
    max_rank = max(df[rank_orig_col].max(), df[rank_dgen_col].max())
    ax.plot([min_rank, max_rank], [min_rank, max_rank], 'r--', alpha=0.6)
    

if len(categories) < len(axes):
    for j in range(len(categories), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("mcg_paper_table_with_ModelShots.csv")


categories = ["humanities", "socialsciences", "stem", "others", "avg"]

# D-GEN - Original difference
for cat in categories:
    orig_col = f"mmlu_original_{cat}_accuracy"
    dgen_col = f"mmlu_dgen_{cat}_accuracy"
    diff_col = f"diff_{cat}"
    

    if orig_col in df.columns and dgen_col in df.columns:
        df[diff_col] = df[dgen_col] - df[orig_col]
    else:
        print(f"'{orig_col}' or '{dgen_col}' not found")
```

```python
category_stats = []

for cat in categories:
    diff_col = f"diff_{cat}"
    if diff_col not in df.columns:
        continue  
    
    desc = df[diff_col].describe() 
    
    category_stats.append([
        cat,
        desc["mean"],
        desc["std"],
        desc["min"],
        desc["25%"],
        desc["50%"],
        desc["75%"],
        desc["max"]
    ])

df_stats = pd.DataFrame(
    category_stats,
    columns=["Category", "Mean", "Std", "Min", "25%", "Median", "75%", "Max"]
)

df_stats = df_stats.round(4)

print(df_stats)
```
